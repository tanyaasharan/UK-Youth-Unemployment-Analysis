```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge

def bayesian_imputation(df, random_state=0):

    # Separate numeric and non-numeric columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    non_numeric_cols = df.select_dtypes(exclude=['float64', 'int64']).columns

    # Store non-numeric data
    non_numeric_data = df[non_numeric_cols]

    # Initialize the Bayesian Ridge imputer
    imputer = IterativeImputer(estimator=BayesianRidge(), random_state=random_state)

    # Impute the numeric data
    imputed_numeric_data = imputer.fit_transform(df[numeric_cols])

    # Convert the imputed array back to a DataFrame
    imputed_df = pd.DataFrame(imputed_numeric_data, columns=numeric_cols, index=df.index)

    # Convert all columns to integers (after rounding)
    imputed_df = imputed_df.round(0).astype(int)

    # Combine with non-numeric data
    imputed_full_df = pd.concat([non_numeric_data, imputed_df], axis=1)

    return imputed_full_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def pca_variance_plot(dataframe, drop_column='Region'):

    # Drop non-numeric column
    features = dataframe.drop(columns=[drop_column])

    # Scale features (i.e. mean = 0 and variance = 1 to avoid bias)
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Apply PCA
    pca = PCA()
    pca_components = pca.fit_transform(scaled_features)

    # Plot cumulative explained variance
    plt.figure(figsize=(10,6))
    plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
             pca.explained_variance_ratio_.cumsum(),
             marker='o', linestyle='--')
    plt.title('Cumulative Explained Variance by PCA Components')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.grid(True)
    plt.show()

    return pca
```

```python
def pca(dataframe, n_components, drop_column='Region'):

    # Drop non-numeric column
    features = dataframe.drop(columns=[drop_column])

    # Scale features
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)

    # Apply PCA with n components
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(features_scaled)

    # Create a DataFrame for PCA components
    component_names = [f'PC{i+1}' for i in range(n_components)]
    pca_df = pd.DataFrame(data=principal_components, columns=component_names)

    # Add the 'Region' back for reference
    pca_df[drop_column] = dataframe[drop_column].reset_index(drop=True)


    return pca_df, pca
```

```python
def elbow_method(pca_df, max_clusters=10, drop_column='Region'):

    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt

    # Drop non-PCA columns (i.e categorical features)
    features = pca_df.drop(columns=[drop_column])

    inertias = []

    for k in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(features)
        inertias.append(kmeans.inertia_)

    # Plot the elbow curve
    plt.figure(figsize=(8,6))
    plt.plot(range(1, max_clusters+1), inertias, marker='o', linestyle='--')
    plt.title('Elbow Method For Optimal k')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia (within-cluster sum of squares)')
    plt.grid(True)
    plt.show()
```

```python
from sklearn.cluster import KMeans

def kmeans_clustering(pca_df, n_clusters=3, drop_column='Region', plot=True):

    # Select only PCA features (drop Region)
    features = pca_df.drop(columns=[drop_column])

    # Apply KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(features)

    # Add cluster labels to the dataframe
    pca_df_clustered = pca_df.copy()
    pca_df_clustered['Cluster'] = cluster_labels

    # Plot if it's 2D
    if plot and features.shape[1] == 2:
        plt.figure(figsize=(8, 6))
        scatter = plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=cluster_labels, cmap='viridis', s=50)
        plt.title('KMeans Clustering (2D PCA)')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.colorbar(scatter, label='Cluster')
        plt.grid(True)
        plt.show()

    return pca_df_clustered, kmeans
```

```python
from sklearn.preprocessing import StandardScaler
import umap
import pandas as pd
import matplotlib.pyplot as plt

def apply_umap_projection(df, drop_column='Region', n_neighbors=15, min_dist=0.1, random_state=42):
    # Drop non-numeric
    features = df.drop(columns=[drop_column])

    # Scale
    scaler = StandardScaler()
    scaled = scaler.fit_transform(features)

    # Apply UMAP
    reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, random_state=random_state)
    embedding = reducer.fit_transform(scaled)

    # Create DataFrame
    umap_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
    umap_df[drop_column] = df[drop_column].values
    return umap_df
```

```python
from sklearn.cluster import KMeans

def cluster_umap(df, drop_column='Region', n_clusters=3):
    features = df.drop(columns=[drop_column])
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    df['Cluster'] = kmeans.fit_predict(features)
    return df, kmeans
```

```python
from sklearn.linear_model import BayesianRidge
import numpy as np
import pandas as pd

def predict_unemployment_2030(df):
    # Define training and prediction years
    X = np.array([[2011], [2021]])
    year_2030 = np.array([[2030]])

    # Safe log function
    def safe_log(x): return np.log(x) if x > 0 else np.nan

    # Log-transform relevant columns
    df['log_Unemp_Total_2011'] = df['Unemployed: Total_2011'].apply(safe_log)
    df['log_Unemp_Total_2021'] = df['Unemployed_2021'].apply(safe_log)
    df['log_Unemp_Male_2011'] = df['Unemployed: Male_2011'].apply(safe_log)
    df['log_Unemp_Male_2021'] = df['Unemployed: Male_2021'].apply(safe_log)
    df['log_Unemp_Female_2011'] = df['Unemployed: Female_2011'].apply(safe_log)
    df['log_Unemp_Female_2021'] = df['Unemployed: Female_2021'].apply(safe_log)

    # Drop rows with any missing logs
    required = [
        'log_Unemp_Total_2011', 'log_Unemp_Total_2021',
        'log_Unemp_Male_2011', 'log_Unemp_Male_2021',
        'log_Unemp_Female_2011', 'log_Unemp_Female_2021'
    ]
    df = df.dropna(subset=required).copy()

    # Predictor function
    def predict_log_value(log_2011, log_2021):
        y = np.array([log_2011, log_2021])
        model = BayesianRidge()
        model.fit(X, y)
        return np.exp(model.predict(year_2030)[0])  # inverse log

    # Apply predictions
    df['Unemployed_2030'] = df.apply(lambda row: predict_log_value(row['log_Unemp_Total_2011'], row['log_Unemp_Total_2021']), axis=1)
    df['Unemployed: Male_2030'] = df.apply(lambda row: predict_log_value(row['log_Unemp_Male_2011'], row['log_Unemp_Male_2021']), axis=1)
    df['Unemployed: Female_2030'] = df.apply(lambda row: predict_log_value(row['log_Unemp_Female_2011'], row['log_Unemp_Female_2021']), axis=1)

    return df
```
